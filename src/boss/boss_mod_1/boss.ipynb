```python
import pandas as pd
```

```python
# carregando dataframe
dt = pd.read_csv('pokedex_(update_05.20).csv')
dt
```

```python
# Remoção de duplicatas
dt.duplicated().any()

# O DataFrame não possui valores duplicados!
```

```python
# exibindo colunas existentes para melhor visualização
dt.columns
```

```python
# Remover colunas irrelevantes
dt = dt.loc[:,['pokedex_number','name', 'status', 'species', 'type_1', 'type_2', 'height_m', 'weight_kg', 'hp', 'attack', 'defense', 'speed', ]]
dt
```

```python
# tratar valores ausentes
dt.isnull().sum()

# coluna type_2 e coluna weight_kg possui valores ausentes
```

```python
# Excluindo valores nulos presentes no DataFrame
dt = dt.dropna()
dt
```

```python
# verificando se os valores ausentes foram deletados
dt.isnull().sum()
```

```python
# Filtrar pokemons proibidos que sejam das categorias:
# Lendários e sub-lendários
# míticos
# pokémon com mega evolução

df = dt[~dt['status'].isin(['Legendary', 'Semi Legendary']) & ~dt['name'].str.contains('mega', case=False)]
df

# Dataframe finalizado
```
