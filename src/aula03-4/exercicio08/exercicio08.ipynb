#### Exercicio 08
✦ Crie mais 2 dataframes, de séries e anime/desenhos que você gosta, com as mesmas colunas acima.
✦ Salve os 3 dataframes em 'csv', 'json' e 'html' e nos envie os arquivos.
✦ Concatene horizontalmente os 3 dataframes, ficando 9 colunas no final.

```python
nome = ["Fallen angels", "O Fabuloso Destino de Amélie Poulain", "princesa Mononoke", "Ladrões de Bicicleta", "Blade Runner"]
data = [1995, 2001, 1997, 1948, 1982]
dt_filmes = pd.DataFrame({'nome':nome, 'Data':data})
dt_filmes['Nota'] = [10, 8, 9, 10, 10]
```

```python
series = {
    'nome': ['Sopranos', 
               'Breaking Bad',
               'Dark',
               'Game of Thrones',
               'Prison Break'],
    'Data': [2000, 2000, 2000, 2000, 2000],
    'Nota': [10, 10, 9, 8, 6]
}

desenhos = {
    'nome': ['Viva: a vida é uma festa',
                'Castelo Animado',
                'Viagem de Chihiro',
                'Rei Leão',
                'Ratatuille'],
    'Data': [2001, 2001, 2001, 2001, 2001],
    'Nota': [10, 9, 7, 8, 10]
}

dt_series = pd.DataFrame(series)
dt_desenhos = pd.DataFrame(desenhos)
```

```python
dt_series
```

```python
dt_filmes.to_csv('filmes.csv')
dt_filmes.to_json('filmes.json')
dt_filmes.to_html('filmes.html')

dt_series.to_csv('series.csv')
dt_series.to_json('series.json')
dt_series.to_html('series.html')

dt_desenhos.to_csv('desenhos.csv')
dt_desenhos.to_json('desenhos.json')
dt_desenhos.to_html('desenhos.html')
```

```python
dt_concat =  pd.concat([dt_filmes, dt_series, dt_desenhos], axis = 1)
dt_concat
```
